```python
# import dependencies
import pandas as pd
import os
```

```python
# getting all csv files
print(f'# Loading the files...')

pwd = os.getcwd()
pwd = pwd + '/Resources/'

for path, folders, files in os.walk(pwd):
    if pwd == path:
        print(files)
        files_names = files

print(f'# Files loading complete')
```

```python
# import data from all csv files
print(f'# Importing data...')

df = []

for x in files_names:
    df.append(pd.read_csv(pwd + x))
    print("Processing data file: ", x)
    
print(f'# Data imported complete')
```

```python
# concatening all dfs in a single Pandas df
df_all = pd.concat([
    df[0], df[1], df[2], df[3], df[4], df[5]])
df_all
```

```python
# replace binary values from gender column with unknown, male, female
df_all.gender = df_all.gender.replace(1, 'Male').replace(2, 'Female').replace(0, 'Unknown')                                                  
df_all
```

```python
# export the data to a csv file
#FIRST create file CitiBikes_2019_summer.csv in output folder
print(f'# Exporting data ... ')
export_csv = df_all.to_csv (r'output/CitiBikes_2019_spring_summer.csv', index = None, header=True) 
print(f'# ...Data exporting complete!')
```
